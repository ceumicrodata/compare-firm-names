```python
import pandas as pd
import numpy as np
from Levenshtein import ratio, jaro_winkler, jaro
from fuzzywuzzy import fuzz
import search
```

```python
df=pd.read_csv("input/foreign_sample.csv")
df['score'] = pd.Series( 0 , index = df.index)
df.Partner = df.Partner.fillna('')
df.Name = df.Name.fillna('')
```

```python
i = 0
while i < len(df):
    
    string1 = search.normalize_text(df.Partner[i])
    string2 = search.normalize_text(df.Name[i])
    print(i)
    leven = round(ratio(string1,string2)*100)
    jaro2 =  round(jaro_winkler(string1,string2, 2.0)*100)
    jaro1 = round(jaro_winkler(string1,string2, 1.5)*100)
    fuzz_part = fuzz.partial_ratio(string1, string2)
    fuzz_token = fuzz.token_set_ratio(string1, string2)
    avg_token_jaro = (fuzz_token + jaro2) / 2
    avg_part_jaro = (fuzz_part + jaro2) / 2
    print([string1, string2])
    print(['leven', leven])
    print(['jaro-winkler 2', jaro2])
    print(['jaro-winkler 1.5', jaro1])
    #better with longer strings
    print(['fuzz_partial', fuzz_part])
    print(['fuzz_token_set_ratio', fuzz_token])
    #fuzz better with prefixes in the beginning like sc.
    print(['avg_token_set-jaro', avg_token_jaro])
    print(['fuzz_partial-jaro', avg_part_jaro])
    i = i+1
    

```
